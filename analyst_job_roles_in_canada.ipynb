Exploratory Data Analysis

```python
plt.figure(figsize=(15, 6))
sns.countplot(x="Job Title", data=df, palette="viridis",  hue="Job Title")
plt.xlabel("Job Title")
plt.ylabel("Count")
plt.title("Distribution of Job Titles")
plt.xticks(rotation=70)
plt.show()
```

```python
group_by1=df.groupby("Position")[["Min_Salary","Max_Salary"]].mean()
```
